```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("dataset_2.csv")
```

```python
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

X = df[['Height', 'Age', 'Gender']]
y = df['Weight']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=30
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Model Intercept:", model.intercept_)
print("Model Coefficients:", model.coef_)
print("Mean Squared Error (Test):", mean_squared_error(y_test, y_pred))
print("R² Score (Test):", r2_score(y_test, y_pred))

comparison = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred.round(2)})
print("\nPrediction Comparison (Test Set):\n", comparison)
```
